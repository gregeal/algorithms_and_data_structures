  # Question One: Check Duplicates in Tree

  Given the `root` of a binary tree, check whether it is contains a duplicate value. If a duplicate exists, return the duplicate value. If there are multiple duplicates, return the one with the closest distance to the root. If no duplicate exists, return -1.

#### Starter Code for Question 1

````python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

from collections import deque
from typing import Optional, List

def build_tree_from_list(values: List) -> Optional[TreeNode]:
    """
    Helper function to build a binary tree from level-order list representation.
    None values in the list represent missing nodes.
    """
    if not values:
        return None
    
    root = TreeNode(values[0])
    queue = deque([root])
    i = 1
    
    while queue and i < len(values):
        node = queue.popleft()
        
        # Add left child
        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1
        
        # Add right child
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1
    
    return root

def is_duplicate(root: Optional[TreeNode]) -> int:
    """
    Find the first duplicate value in a binary tree using level-order traversal.
    Returns the duplicate closest to the root, or -1 if no duplicates exist.
    
    Time Complexity: O(n) where n is the number of nodes
    Space Complexity: O(n) for the queue and set
    """
    if not root:
        return -1
    
    # Use a set to track values we've seen
    seen = set()
    
    # Use a queue for level-order (BFS) traversal
    # This ensures we check nodes level by level, closest to root first
    queue = deque([root])
    
    while queue:
        node = queue.popleft()
        
        # Check if we've seen this value before
        if node.val in seen:
            return node.val
        
        # Mark this value as seen
        seen.add(node.val)
        
        # Add children to queue for next level
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    
    # No duplicates found
    return -1
````

````python
# Test cases
print("Example 1:")
root1 = build_tree_from_list([1, 2, 2, 3, 5, 6, 7])
print(f"Input: [1, 2, 2, 3, 5, 6, 7]")
print(f"Output: {is_duplicate(root1)}")
print(f"Expected: 2\n")

print("Example 2:")
root2 = build_tree_from_list([1, 10, 2, 3, 10, 12, 12])
print(f"Input: [1, 10, 2, 3, 10, 12, 12]")
print(f"Output: {is_duplicate(root2)}")
print(f"Expected: 10\n")

print("Example 3:")
root3 = build_tree_from_list([10, 9, 8, 7])
print(f"Input: [10, 9, 8, 7]")
print(f"Output: {is_duplicate(root3)}")
print(f"Expected: -1")
````

````python
# Your answer here
from typing import List

def first_duplicate(nums: List[int]) -> int:
    index = 0
    for num in nums:
        if num in nums[index + 1:]:
            return num
        index += 1
    return -1

first_duplicate([6, 2, 1, 3, 3, 5, 2, 1]) \
,first_duplicate([1, 3, 2, 8]) \
,first_duplicate([3, 2, 9, 23, 45, 1, 7, 7])
````


-   Explain why their solution works in your own words.


### Why the Solution Works

The solution works by iterating through each element in the list and checking if that element appears anywhere in the remaining portion of the list (after the current position). This is accomplished using list slicing `nums[index + 1:]` which creates a sublist containing all elements after the current index.

For each element at position `index`, the code checks if that same value exists in the rest of the list. If it does, this means we've found a duplicate, and since we're scanning from left to right, this is the first element that has a duplicate. The function immediately returns this value.

If we complete the entire loop without finding any duplicates, the function returns -1 to indicate no duplicates exist.

The key insight is that by checking only the portion of the list that comes AFTER the current element, we ensure we find duplicates and we find them in the order they first appear.

### Time and Space Complexity

**Time Complexity: O(nÂ²)**

The outer loop iterates through all n elements in the list, which is O(n). For each iteration, the `in` operator checks if the current element exists in the remaining slice of the list. In the worst case, this check examines the entire remaining portion of the list. The `in` operator for lists performs a linear search, which is O(n) in the worst case.

Therefore, the overall time complexity is O(n) Ã— O(n) = O(nÂ²).

In the worst case scenario (when there are no duplicates or the duplicate is at the very end), the algorithm will need to compare nearly every element with every other element.

**Space Complexity: O(n)**

The primary space used is:
1. The input list `nums` which takes O(n) space
2. The slice operation `nums[index + 1:]` creates a new list, which in the worst case (when index = 0) contains n-1 elements, so O(n) space
3. The variable `num` and `index` use O(1) space

The dominant factor is the list slicing, which creates a copy of a portion of the original list. Therefore, the overall space complexity is O(n).

def first_duplicate_optimized(nums: List[int]) -> int:
